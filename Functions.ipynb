6. Explain the concept of generators in Python and how they are defined?

    - Generators are special functions that return an iterator using the yield keyword instead of return. They allow you to iterate over a sequence of values lazily — meaning they generate items one at a time only when needed.

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
for value in gen:
    print(value)
```

7. What are the advantages of using generators over regular functions?


  -  Memory Efficient: They don’t store the entire sequence in memory.

  - Lazy Evaluation: Values are computed on demand.

  - Faster for large datasets: Especially when looping through massive data.

Example Comparison:

Using a generator:

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for number in count_up_to(5):
    print(number)
```

10. What is the difference between map(), reduce(), and filter() functions in Python?

map()
   - Purpose: Applies a given function to each item in an iterable (like a list).

   -  Returns: A map object (an iterator) which can be converted to a list.

   

```python
result = map(lambda x: x + 1, [1, 2, 3])
print(list(result))
```

filter()

  - Purpose: Filters out the items in an iterable for which the given function returns True.

  - Returns: A filter object (an iterator), which can be converted to a list.

```python
result = filter(lambda x: x > 2, [1, 2, 3])
print(list(result))
```

reduce()

  - Purpose: Repeatedly applies a function to the iterable and reduces it to a single cumulative value.

  - Returns: A single final value (not an iterator).

Note: reduce() is part of the functools module and must be imported.

```python
from functools import reduce
result = reduce(lambda x, y: x + y, [1, 2, 3])
print(result)
```

Function to check if a given number (from 1 to 200) is prime


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Test from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

5. create aniterator class in python that generates the Fibonacci sequence up to a specified number of terms.

Here is a Python iterator class that generates the Fibonacci sequence up to a specified number of terms:

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms  # Total number of terms to generate
        self.count = 0              # Counter for the number of terms generated
        self.a = 0                  # First term
        self.b = 1                  # Second term

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. write a generator function in python that yields the power of 2 up to a given exponent.


Here's a Python generator function that yields powers of 2 up to a given exponent:

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage:
for value in powers_of_two(5):
    print(value)
```

Explanation:

The class keeps track of the current term count and the last two Fibonacci numbers.

It implements the __iter__() and __next__() methods to make it an iterator.

The sequence is generated until the desired number of terms is reached.

7. implement a generator function that reads a file line by line  and yields each line as  a string

Here's a Python generator function that reads a file line by line and yields each line as a string:

```python
# Sample file
with open('sample.txt', 'w') as f:
    f.write("Line 1 Priya Garhewal\n")
    f.write("Line 2 Physics Wallah\n")
    f.write("Line 3 Data Analytics Course\n")

# Reading file using generator
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # remove newline

for line in read_file_line_by_line('sample.txt'):
    print(line)
```

9. write a python program that uses "map()" to convert a list of temperatures from Celsius to  Fahrenheit

```python
# Celsius to Fahrenheit conversion formula: F = (C * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to convert to Fahrenheit
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Print the result
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

10. create a python program that uses "filter()" to remove all the vowels from a given string

```python
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    # Use filter with lambda to exclude vowels
    result = filter(lambda ch: ch not in vowels, input_string)
    return ''.join(result)

# Example usage
text = "Hello, World!"
filtered_text = remove_vowels(text)
print(filtered_text)
```
